```python
import pandas as pd
import sys
assert sys.version_info >= (3, 5)

# Scikit-Learn ≥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"

# Common imports
import numpy as np
import os
import matplotlib as mpl
mpl.use('TkAgg') 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
plt.ion()
```

```python
def load_NSIN_data():
    df = pd.read_csv('../datasets/deviceCoordinates.csv')

    return df
```

```python
coordinates = load_NSIN_data()

coordinates.head(n=5)
```

```python
coordinates = coordinates.drop('Unnamed: 1', inplace=False, axis=1)
coordinates = coordinates.drop('Unnamed: 2', inplace=False, axis=1)
coordinates = coordinates.drop('Unnamed: 3', inplace=False, axis=1)
coordinates
```

```python
x = coordinates['Latitude']
y = coordinates['Longitude']
z = coordinates['Altitude']

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(x,y,z,c='b', marker='o')

ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
ax.set_zlabel('Altitude')

plt.show()

```

```python
actual_latitude = x.to_numpy()
actual_longitude = y.to_numpy()
actual_altitude = z.to_numpy()


predicted_latitude = coordinates['Latitude(Predicted)'].to_numpy()
predicted_longitude= coordinates['Longitude(Predicted)'].to_numpy()
predicted_altitude= coordinates['Altitude(Predicted)'].to_numpy()

latitude_diff = actual_latitude - predicted_latitude
longitude_diff = actual_longitude - predicted_longitude
altitude_diff = actual_altitude - predicted_altitude
threat_id = coordinates['Threat Id'].to_numpy()
```

```python
fig = plt.figure(figsize=(8,6))
plt.plot(actual_latitude,predicted_latitude, 'rx')
plt.xlabel('actual_latitude')
plt.ylabel('predicted_latitude')
plt.title('Latitude Difference')
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
plt.plot(actual_longitude,predicted_longitude, 'rx')
plt.xlabel('actual_longitude')
plt.ylabel('predicted_longitude')
plt.title('Longitude Difference')
plt.show()
```

```python
fig = plt.figure(figsize=(8,6))
plt.plot(actual_altitude,predicted_altitude,'rx')
plt.xlabel('actual_altitude')
plt.ylabel('predicted_altitude')
plt.title('Altitude Difference')
plt.show()
```

Predictions are near perfect as the graph is making a linear relationship with actual vs predicted.
