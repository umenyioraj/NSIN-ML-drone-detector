```python
import pandas as pd
import sys
assert sys.version_info >= (3, 5)

# Scikit-Learn ≥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"

# Common imports
import numpy as np
import os
import matplotlib as mpl
mpl.use('TkAgg') 
import matplotlib.pyplot as plt
plt.ion()

from datetime import datetime
from scipy.interpolate import BSpline
from scipy.interpolate import splrep, splev, CubicSpline
```

```python
def load_BirdData(filename):
    df = pd.read_csv(filename)
    
    return df
```

```python
data = load_BirdData('../datasets/BirdData4.txt')

data.head()
```

```python
data.columns = data.columns.str.strip() #There's extra white space in the data
data.columns = data.columns.str.replace("'", "") #The columns already contain quotes we need to remove 
#data.fillna('0', inplace=True)

Time = data['X'].to_numpy()
Height = data['Y'].to_numpy()

graph_data = {'Time':[], 'Height':[]}

used_list = []
pairs = []
for i in range(0, len(Time)):
    if Time[i] not in used_list:
        pairs.append([Time[i],Height[i]])
        used_list.append(Time[i])
    else:
        next

pairs = sorted(pairs)
for item in pairs:
    graph_data['Time'].append(item[0])
    graph_data['Height'].append(item[1])

t0 = graph_data['Time'][0]
graph_data['Time'][0] = 0
for i in range(1, len(graph_data['Time'])):
    graph_data['Time'][i] = graph_data['Time'][i] - t0

```

```python
# Spline Interpolation of the Trajectory of Multiple Bird Graphs (run very first block with libraries and this block)

listOfBirdDataFiles = ['../datasets/BirdData1.txt','../datasets/BirdData2.txt','../datasets/BirdData3.txt','../datasets/BirdData4.txt']

for file in listOfBirdDataFiles:

    # get title for graph
    graph_name = (((file.split('/'))[2]).split('.'))[0]

    # load data from csv to dataframe
    data = load_BirdData(file)

    data.head()

    data.columns = data.columns.str.strip() #There's extra white space in the data
    data.columns = data.columns.str.replace("'", "") #The columns already contain quotes we need to remove 

    # graph Time and Height data
    Time = data['X'].to_numpy()
    Height = data['Y'].to_numpy()

    graph_data = {'Time':[], 'Height':[]}

    # parsing and skipping repeated time values (repeated time values not accepted by spline function)
    used_list = []
    pairs = []
    for i in range(0, len(Time)):
        if Time[i] not in used_list:
            pairs.append([Time[i],Height[i]])
            used_list.append(Time[i])
        else:
            next

    # sorting values by time in case not sorted already (non-ascending x values not accepted by spline function)
    pairs = sorted(pairs)
    for item in pairs:
        graph_data['Time'].append(item[0])
        graph_data['Height'].append(item[1])

    # converting Time values to time from the start time, t0 (in miliseconds)
    t0 = graph_data['Time'][0]
    graph_data['Time'][0] = 0
    for i in range(1, len(graph_data['Time'])):
        graph_data['Time'][i] = graph_data['Time'][i] - t0

    x = graph_data['Time']
    y = graph_data['Height']

    # Find spline representation
    tck = splrep(x, y)

    # Evaluate spline at new points
    xnew = np.linspace(min(x), max(x), 1000)
    ynew = splev(xnew, tck)

    # Plotting
    plt.figure(figsize=(20,8))
    plt.plot(xnew, ynew, c='r', zorder=1)
    plt.scatter(x, y, marker='.', c='b', zorder=2)
    plt.title(f'Spline Interpolation of {graph_name}')
    plt.xlabel('Time from Detection (milliseconds)')
    plt.ylabel('S(x): Height (meters)')
    plt.show()
```
